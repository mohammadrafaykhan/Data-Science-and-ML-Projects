### 3.7) TRANSFORMATION OF CATEGORICAL VARIABLES

#### We will transform the below categorical variables by assigning them dummy values 

### 3.7.1) Fuel_Type

```python
data['Fuel_Type'].unique()
```

```python
def FUELTYPE(val):
    if val == "Diesel":     
        return "0"
    elif val == "Petrol" :           
        return "1"
    else:
        return "2"

data["Fuel_Type"] = data["Fuel_Type"].apply(FUELTYPE)
```

```python
data["Fuel_Type"].unique()
```

#### Changing datatype from Object to Int

```python
data["Fuel_Type"] = data['Fuel_Type'].astype('int64')
```

```python
data['Fuel_Type'].dtypes
```

### 3.7.2) Color

```python
data['Color'].unique()
```

```python
def COLOR(val):
    if val == "Blue":     
        return "0"
    elif val == "Silver":           
        return "1"
    elif val == "Black":           
        return "2"
    elif val == "White":           
        return "3"
    elif val == "Grey":
        return "4"
    elif val == "Red":
        return "5"
    elif val == "Green":
        return "6"
    elif val == "Yellow":
        return "7"
    elif val == "Violet":
        return "8"
    elif val == "Beige":
        return "9"

data["Color"] = data["Color"].apply(COLOR)
```

```python
data['Color'].unique()
```

#### Changing datatype from Object to Int

```python
data["Color"] = data['Color'].astype('int64')
```

```python
data['Color'].dtypes
```

### 3.7.3) Mfg_Year

```python
data['Mfg_Year'].unique()
```

```python
def MFGYEAR(val):
    if val == 1998:     
        return "0"
    elif val == 1999:           
        return "1"
    elif val == 2000:           
        return "2"
    elif val == 2001:           
        return "3"
    elif val == 2002:
        return "4"
    elif val == 2003:
        return "5"
    elif val == 2004:
        return "6"
 
data["Mfg_Year"] = data["Mfg_Year"].apply(MFGYEAR)
```

```python
data['Mfg_Year'].unique()
```

### 4.1)Pearson Coorelation Matrix for feature selection in numeric data

### 4.3) ANOVA F score for feature selection in nominal categorical data 

#### F score method is used when the data is categorical(nominal) and numeric

#### Extracting Categorical(nominal) features

```python
X=data[['Automatic', 'Color', 'Met_Color', 'Fuel_Type', 'Mfg_Guarantee',
       'BOVAG_Guarantee', 'ABS', 'Airbag_1', 'Airbag_2', 'Airco',
       'Automatic_airco', 'Boardcomputer', 'CD_Player', 'Central_Lock',
       'Powered_Windows', 'Power_Steering', 'Radio', 'Mistlamps',
       'Sport_Model', 'Backseat_Divider', 'Metallic_Rim', 'Radio_cassette',
       'Tow_Bar']]
y=data['Price']
```

```python
import sklearn.feature_selection as fs
fscore = fs.SelectKBest(k='all')
fscore.fit(X, y)
names = X.columns.values[fscore.get_support()]
scores = fscore.scores_[fscore.get_support()]
names_scores = list(zip(names, scores))
df2 = pd.DataFrame(data = names_scores, columns=
  ['Features','F_Scores'])
df2_sorted = df2.sort_values(['F_Scores','Features'], ascending =
  [False, True])
print(df2_sorted)
```

#### Based on high F scores we would be taking Automatic_airco,Boardcomputer and CD_Player

### STEP 5) MODEL TRAINING

#### Selected features with the output variable

```python
x1=data[['Age_08_04','KM','Weight','Automatic_airco', 'Boardcomputer', 'CD_Player']]
y1=data['Price']
```

#### Splitting the data in train and test

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size = 0.30 ,random_state = 2)
```

#### Applying the model and computing R2,RMSE and CV

```python

model = linear_model.LinearRegression()
model.fit(x1_train, y1_train)
y1_pred = model.predict(x1_test)

MSE = np.square(np.subtract(y1_test,y1_pred)).mean() 
RMSE = math.sqrt(MSE)
r2 = r2_score(y1_test, y1_pred)
cv_linear = cross_val_score(model,x1_train,y1_train)
print('RMSE for the data:',RMSE)
print('Rsquared for the data:',r2)
print("CV: ", cv_linear.mean())
```
