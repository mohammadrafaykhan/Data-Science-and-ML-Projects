## Olympics Atheletes Dataset 
This is a historical dataset on the modern Olympic Games, including all the Games from Athens 1896 to Rio 2016 from www.sports-reference.com in May 2018. 

Dataset can be downloaded from here: https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results

Following are the columns in the dataset:
- ID - Unique number for each athlete
- Name - Athlete's name
- Sex - M or F
- Age - Integer
- Height - In centimeters
- Weight - In kilograms
- Team - Team name
- NOC - National Olympic Committee 3-letter code
- Games - Year and season
- Year - Integer
- Season - Summer or Winter
- City - Host city
- Sport - Sport
- Event - Event
- Medal - Gold, Silver, Bronze, or NA

# 1) Description of variables/columns/features in the dataset

1)ID:Unique identification of athlete by a number

2)Name:Name of athlete

3)Sex:Gender of athlete

4)Age:Age of athlete

5)Height:Height of athlete

6)Weight:Weight of athlete

7)Team:Country/Team of athlete

8)NOC:Abbreviation of Country/Team of athlete

9)Games:Year and season in which the athlete took part in the Olympics

10)Year:Year in which the athlete took part in the Olympics

11)Season:Season in which the athlete took part in the Olympics

12)City:The city in which Olympics were played

13)Sport:The sports played by the athlete

14)Event:The event in which the athlete participated

15)Medal:Medal recieved or not by the athlete

### In order to select between mean or median we have to look for outliers in the box plot.Nan will already be excluded so even if we remove them using dropna then still the plot will be same so I am not dropping them.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(data=data[["Age","Height","Weight"]])
plt.show()
```

### We have seen that all 3 plots contain outliers.Now we have to find the count of these outliers,compare them with the total data and decide whether to use mean or median for missing values

### 1)Age

#### Method 1: Using Q3,Q1 and IQR

```python
#IQR=Q3-Q1
def outliers_age(Q3,Q1,IQR):
    outliers_lowerbound=Q1-(1.5*IQR)
    outliers_upperbound=Q3+(1.5*IQR)
    print("Lower bound:",outliers_lowerbound)
    print("Upper bound:",outliers_upperbound)
outliers_age(28,21,7)
```

```python
a=data["Age"]
count=0
for i in a:
    if i>38.5 or i<10.5:
        count=count+1
print("total outliers in age are:",count)
percentage_of_ageoutliers=(count/271116)*100 
print("percentage of age outliers are:",percentage_of_ageoutliers)
        
```

#### Method 2: Using Quantile and percentage

```python
top_age = data['Age'].quantile(0.99)
print("Max value for 99 percent data:",top_age)
data[data['Age'] > top_age]
```

```python
bottom_age = data['Age'].quantile(0.01)
print("Min value for 1 percent data:",bottom_age) 
data[data['Age'] <= bottom_age]
```

### 2) Height

```python
#IQR=Q3-Q1
def outliers_height(Q3,Q1,IQR):
    outliers_lowerbound=Q1-(1.5*IQR)
    outliers_upperbound=Q3+(1.5*IQR)
    print("Lower bound:",outliers_lowerbound)
    print("Upper bound:",outliers_upperbound)
outliers_height(183,168,15)
```

```python
a1=data["Height"]
count=0
for i in a1:
    if i>205.5 or i<145.5:
        count=count+1
print("total outliers in height are:",count)
percentage_of_heightoutliers=(count/271116)*100 
print("percentage of height outliers are:",percentage_of_heightoutliers)
```

#### Method 2: Using Quantile and percentage

```python
top_height = data['Height'].quantile(0.99)
print("Max value for 99 percent data:",top_height)
data[data['Height'] > top_height]
```

```python
bottom_height = merged_data['Height'].quantile(0.01)
print("Min value for 1 percent data:",bottom_height) 
merged_data[ merged_data['Height'] <= bottom_height]
```

### 3) Weight

#### Method 1: Using Q3,Q1 and IQR

```python
#IQR=Q3-Q1
def outliers_weight(Q3,Q1,IQR):
    outliers_lowerbound=Q1-(1.5*IQR)
    outliers_upperbound=Q3+(1.5*IQR)
    print("Lower bound:",outliers_lowerbound)
    print("Upper bound:",outliers_upperbound)
outliers_weight(79,60,19)
```

```python
a2=data["Weight"]
count=0
for i in a2:
    if i>107.5 or i<31.5:
        count=count+1
print("total outliers in weight are:",count)
percentage_of_weightoutliers=(count/271116)*100 
print("percentage of weight outliers are:",percentage_of_weightoutliers)
        
```

#### Method 2: Using Quantile and percentage

```python
top_weight = data['Weight'].quantile(0.99)
print("Max value for 99 percent data:",top_weight)
data[data['Weight'] > top_weight]
```

```python
bottom_weight = data['Weight'].quantile(0.01)
print("Min value for 1 percent data:",bottom_weight) 
data[data['Weight'] <= bottom_weight]
```

### We now have an option to use Method 1 or Method 2.I select Method 1.According to it since the percentages obtained for outliers are negligible as compared to the whole data we have an option to take mean or median both.Another reason is that since approximately mean=median it forms a normal distribution so we can take either value.I will be going with median.


Q1. Which country has the most gold medalists?

```python
# add your code answer here 
z=fixed_data[fixed_data['Medal'] == "Gold"]
z.groupby("Medal")["Team"].value_counts()
```

#### Converted series to dataframe,renamed the column and visualized the top 10 countries with most Gold medals.The country with the most Gold medals is clearly shown.

```python
# add your visualization code here
import seaborn as sns

df_Q1=z.groupby("Medal")["Team"].value_counts()
df1_Q1=pd.DataFrame(df_Q1)
df1_Q1.columns = [*df1_Q1.columns[:-1], 'Number of Gold Medals']
g=df1_Q1.reset_index()
g1=g.nlargest(10,columns='Number of Gold Medals')[['Team','Number of Gold Medals']]
sns.set(rc={"figure.figsize":(12, 8)})
sns.barplot(x = 'Team', y = 'Number of Gold Medals', data = g1)
```

#### The team with most Gold medals is United States(2474)
